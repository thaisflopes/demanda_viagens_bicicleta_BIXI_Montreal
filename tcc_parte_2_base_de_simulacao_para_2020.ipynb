```python
#Importando bibliotecas
import pandas as pd
import itertools
from itertools import product
```

### Base com todas as combinações possiveis de 2020

```python
#Criando as listas com as semanas, estações e membro
lista_semana = list(range(0, 32))
estacoes_2020 = pd.read_csv("Stations_2020.csv")
estacoes_2020 = list(estacoes_2020['code'])
lista_membro = [0,1]
```

```python
#Gerando todas as combinações possíveis
combinacoes_2020 = list(itertools.product(lista_semana, estacoes_2020, estacoes_2020, lista_membro))
combinacoes_2020 = pd.DataFrame(combinacoes_2020, columns = ['n_semana','start_station_code','end_station_code','is_member'])
combinacoes_2020['ano'] = 2020
combinacoes_2020['n_semana,inicio,fim,membro'] = combinacoes_2020.apply(lambda x:'%s,%s,%s,%s' % (x['n_semana'], x['start_station_code'],x['end_station_code'],x['is_member']),axis=1)
combinacoes_2020['n_semana,inicio,fim,membro'] = combinacoes_2020['n_semana,inicio,fim,membro'].astype(object)
combinacoes_2020.head()
```

```python
#Baixando o dataset de dados reais de 2020
dados_2020 = pd.read_csv("dados_2020.csv")
dados_2020.head()
```

```python
#Estabelecendo as combinações de 'n_semana,inicio,fim,membro' em dados_2020
dados_2020['n_semana,inicio,fim,membro'] = dados_2020.apply(lambda x:'%s,%s,%s,%s' % (x['n_semana'], x['start_station_code'],x['end_station_code'],x['is_member']),axis=1)
dados_2020['n_semana,inicio,fim,membro'] = dados_2020['n_semana,inicio,fim,membro'].astype(object)
dados_2020.dtypes
```

```python
#Filtrando qtd_viagens
qtd_viagens_2020 = dados_2020.drop(labels=['Unnamed: 0', 'start_station_code', 'end_station_code', 'is_member', 'semana', 'ano', 'mes', 'semana,inicio,fim,membro', 'n_semana'], axis = 1)
qtd_viagens_2020.head()
```

```python
#Juntando combinacoes_2020 e qtd_viagens_2020
simulacao_2020 = pd.merge(combinacoes_2020, qtd_viagens_2020, how='outer', on='n_semana,inicio,fim,membro')
simulacao_2020.head()
```

```python
#Filtrando mes
meses_2020 = dados_2020.drop(labels=['Unnamed: 0', 'start_station_code', 'end_station_code', 'is_member', 'semana', 'ano', 'semana,inicio,fim,membro', 'n_semana,inicio,fim,membro', 'qtd_viagens'], axis = 1)
meses_2020 = meses_2020.drop_duplicates()
meses_2020.head()
```

```python
#Juntando simulacao_2020 e meses_2020
simulacao_2020 = pd.merge(simulacao_2020, meses_2020, how='inner', on='n_semana')
simulacao_2020.head()
```

```python
#Tratando dados nulos
simulacao_2020 = simulacao_2020.fillna(0)
simulacao_2020.head()
```

```python
simulacao_2020.to_csv('simulacao_2020.csv')
```

### Base combinando os registros de 2019 e 2020

```python
#Baixando o dataset de dados reais de 2019
dados_2019 = pd.read_csv("dados_2019.csv")
dados_2019.head()
```

```python
#Alterando o ano para 2020
dados_2019['ano'] = 2020
dados_2019
```

```python
#Estabelecendo as combinações de 'n_semana,inicio,fim,membro' em dados_2019
dados_2019['n_semana,inicio,fim,membro'] = dados_2019.apply(lambda x:'%s,%s,%s,%s' % (x['n_semana'], x['start_station_code'],x['end_station_code'],x['is_member']),axis=1)
dados_2019['n_semana,inicio,fim,membro'] = dados_2019['n_semana,inicio,fim,membro'].astype(object)
dados_2019.dtypes
```

```python
#Juntando dados_2019 e dados_2020
simulacao_2020b = pd.merge(dados_2019, dados_2020, how='outer', on='n_semana,inicio,fim,membro')
simulacao_2020b.head()
```

```python
#Vendo onde tem na
print(simulacao_2020b.isnull().sum())
```

```python
#Substituindo os valores na pelo valor de outra coluna
simulacao_2020b['n_semana_x'] = simulacao_2020b['n_semana_x'].fillna(simulacao_2020b['n_semana_y'])
simulacao_2020b['start_station_code_x'] = simulacao_2020b['start_station_code_x'].fillna(simulacao_2020b['start_station_code_y'])
simulacao_2020b['end_station_code_x'] = simulacao_2020b['end_station_code_x'].fillna(simulacao_2020b['end_station_code_y'])
simulacao_2020b['is_member_x'] = simulacao_2020b['is_member_x'].fillna(simulacao_2020b['is_member_y'])
simulacao_2020b['ano_x'] = simulacao_2020b['ano_x'].fillna(simulacao_2020b['ano_y'])
simulacao_2020b['mes_x'] = simulacao_2020b['mes_x'].fillna(simulacao_2020b['mes_y'])
print(simulacao_2020b.isnull().sum())
```

```python
#Retirando colunas
simulacao_2020b = simulacao_2020b.drop(labels=['Unnamed: 0_x', 'semana_x', 'semana,inicio,fim,membro_x', 'Unnamed: 0_y', 'semana,inicio,fim,membro_y', 'qtd_viagens_x', 'start_station_code_y', 'end_station_code_y', 'is_member_y', 'semana_y', 'ano_y', 'mes_y', 'n_semana_y'], axis = 1)
simulacao_2020b.head()
```

```python
#Tratando dados nulos
simulacao_2020b = simulacao_2020b.fillna(0)
simulacao_2020b.head()
```
